```python
import re

def extract_regular_term_fees(text):
    if not isinstance(text, str):
        return None  # หรือ "" หรือ 0 แล้วแต่ต้องการ

    fees = []
    lines = text.split('\n')

    for line in lines:
        line = line.strip()

        # ตัดทิ้งภาคฤดูร้อนหรือ summer
        line = re.split(r'ภาคฤดูร้อน|summer', line, flags=re.IGNORECASE)[0]

        # Pattern 1: ภาคการศึกษาละ 25,500 บาท
        match = re.search(r'ภาค(การศึกษา|เรียน|เทอม)(ละ)?\s*([\d,]+)\s*บาท', line)
        if match:
            fees.append(int(match.group(3).replace(',', '')))
            continue

        # Pattern 2: 20,000 บาท / ภาคการศึกษา
        match = re.search(r'([\d,]+)\s*บาท\s*/\s*(ภาคการศึกษา|ภาคเรียน|เทอม)', line, re.IGNORECASE)
        if match:
            fees.append(int(match.group(1).replace(',', '')))
            continue

        # Pattern 3: 20,000 บาท ต่อภาคการศึกษา
        match = re.search(r'([\d,]+)\s*บาท?\s*ต่อ\s*(ภาคการศึกษา|ภาคเรียน|เทอม)', line, re.IGNORECASE)
        if match:
            fees.append(int(match.group(1).replace(',', '')))
            continue

        # Pattern 4: 175,000 ต่อภาคการศึกษา (ไม่มีคำว่า "บาท")
        match = re.search(r'([\d,]+)\s*ต่อ\s*(ภาคการศึกษา|ภาคเรียน|เทอม)', line, re.IGNORECASE)
        if match:
            fees.append(int(match.group(1).replace(',', '')))
            continue

        # Pattern 5: 18,000.-/ภาคการศึกษา
        match = re.search(r'([\d,]+)(?:\.\-?)?/ภาค(?:การศึกษา|เรียน|เทอม)', line, re.IGNORECASE)
        if match:
            fees.append(int(match.group(1).replace(',', '')))
            continue

        # Pattern 6: ตลอดหลักสูตร + จำนวนภาค
        match = re.search(r'([\d,]+)\s*บาท\s*ต่อ.*?เรียน\s*(\d+)\s*ภาค', line, re.IGNORECASE)
        if match:
            total = int(match.group(1).replace(',', ''))
            term_count = int(match.group(2))
            fees.append(total // term_count)
            continue

        # Pattern 7: ตลอดหลักสูตร → หาร 8
        match = re.search(r'([\d,]+)\s*บาท\s*ตลอดหลักสูตร', line, re.IGNORECASE)
        if match:
            total = int(match.group(1).replace(',', ''))
            fees.append(total // 8)
            continue

        # Pattern 8: เทอมแรก/เทอมต่อไป
        match = re.findall(r'เทอม(?:แรก|ต่อไป)[^\d]*([\d,]+)\s*บาท', line)
        for m in match:
            fees.append(int(m.replace(',', '')))

        # Pattern 9: 72,610 บาท (ไม่มีคำว่า "ต่อ" หรือ "ตลอด")
        match = re.search(r'^([\d,]+)\s*บาท\s*$', line)
        if match:
            fees.append(int(match.group(1).replace(',', '')))
            continue

    # ถ้าเจอค่า → คืน list (หรือจะเลือกค่าตัวเดียวก็ได้)
    if fees:
        return fees[0]

    # ไม่ตรง pattern ใดเลย → คืนข้อความต้นฉบับ
    return text
```

```python
df['ค่าใช้จ่าย'] = df['ค่าใช้จ่าย'].apply(extract_regular_term_fees)
```

```python
def adjust_fee(row):
    fee = row['ค่าใช้จ่าย']
    course_type = str(row['ประเภทหลักสูตร'])

    # แปลงค่าให้เป็นตัวเลขเดี่ยวก่อน (จาก list หรือ string)
    if isinstance(fee, list) and fee:
        fee = fee[0]  # เอาค่าแรกจาก list
    elif isinstance(fee, str):
        fee = fee.replace(',', '').strip()
        if fee.isdigit():
            fee = int(fee)
        else:
            return fee  # ถ้าแปลงไม่ได้ ปล่อยค่าเดิม

    if isinstance(fee, (int, float)):
        if fee > 100000 and 'นานาชาติ' not in course_type:
            return fee // 8

    return fee
```

```python
df['ค่าใช้จ่ายที่ปรับแล้ว'] = df.apply(adjust_fee, axis=1)
```
