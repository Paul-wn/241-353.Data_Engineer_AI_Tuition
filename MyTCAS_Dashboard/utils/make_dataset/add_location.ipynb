```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
def split_coordinates(row):
    if isinstance(row['latitude'], str) and ',' in row['latitude']:
        lat, lon = row['latitude'].split(', ')
        return pd.Series([float(lat), float(lon)])
    return pd.Series([row['latitude'], row['longitude']])

# ใช้ฟังก์ชันกับ DataFrame
df[['latitude', 'longitude']] = df.apply(split_coordinates, axis=1)
```

```python
df = df[df['คณะ'] != 'คณะครุศาสตร์อุตสาหกรรม']
```

```python
df.to_csv('data_via_location.csv' , index = False)
```
